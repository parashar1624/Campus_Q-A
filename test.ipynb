```python
from google import genai   #calling LLM using API Key.

client = genai.Client()

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="Weather in New York is 20 degrees Celsius with clear skies. What is the weather like in San Francisco?",
)

print(response.text)
```
